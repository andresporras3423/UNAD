Reconocimiento de las herramientas

1) ¿Qué es Jupyter y cómo puede serle útil a un científico de datos? 
Jupyter es un programa de código abierto que permite crear y compartir código en diversos lenguajes de programación. Su nombre originalmente viene de la combinación de tres lenguajes que soporta la plataforma (Julia, Python y R). Una de las principales características de Jupiter son sus notebooks, este permite crear y compartir documentos que contienen código ejecutable, gráficos generados por código, imágenes y texto explicativo.
Para un científico de datos Jupyter ofrece los siguientes beneficios:
    1. ejecución de fragmentos código a través de celdas de notebook, con resultados inmediatos de la ejecución.
    2. Documentar el código con texto explicativo usando markdown.
    3. Visualización de gráficos generados por código, por ejemplo con herramientas como Matplotlib, Seaborn y Plotly.
    4. Parametrizar algoritmos para experimentar con diferentes modelos.
    5. Registrar las acciones realizadas, tales como resultados obtenidos o celdas ejecutadas.
    6. Compartir Jupyter notebooks con propositos de colaboración en el desarrollo de proyectos en ciencia de datos.
    7. Para educadores en ciencia de datos,  Jupyter permite la creación materiales interactivos y ejercicios prácticos.
2) ¿Qué es una celda markdown? 

En los Jupyter notebooks, las celdas markdown son un tipo de celda que nos permiten escribir lenguaje de marcado Markdown. Este es un lenguaje de marcado diseñado para ser más ligero y sencillo que alternativas similares como lo es HTML. Markdown permite darle formato al texto usando encabezados, listas, enlaces, imágenes, entre otros elementos.

3) ¿Qué es un paquete (o librería) en Python? 

En python un paquete una carpeta que contiene una colección de módulos, cada uno con funciones y clases, todos relacionados y encapsulados para ser usados en tareas específicas. Estos paquetes están diseñados para ser fácilmente compartidos, lo que facilita la colaboración y rehusar soluciones ya existentes para problemas frecuentes.

4) Describa para qué sirven cada una de las siguientes librerías 

    1. Pip: Es un sistema de paquetes utilizado en python para administrar e instalar paquetes de terceros.
    2. NumPy: librería utilizada frecuentemente en python para la computación científica y el análisis numérico. Tiene una amplia variedad de funciones matemáticas y es la base para muchos otros paquetes y librerías para la ciencia de datos.
    3.  Pandas: librería que ofrece un amplio rango de estructuras de datos y otras herramientas hacer y manipular datos de manera eficiente.
    4. Seaborn: librería para visualización de datos, diseñada específicamente para hacer gráficos estadísticos. Está construida sobre Matplotlib,  
    5. matplotlib: Librería para visualización de datos de forma versátil. Permite visualizar datos multidimensionales, personalizar gráficos y además está diseñado para integrarse fácilmente a otros librerías populares como Numpy y Pandas.
    6. Scikit-learn: librería para hacer machine learning. Permite hacer agrupamiento, clasificación, regresión, selección de características, variedad de algoritmos, preprocesamiento de datos, evaluación de modelos, validación cruzada, entre otros.
    7. Requests: Librería para hacer solicitudes HTTP. Funciona de manera nativa con la arquitectura RESTFUL pero también puede usarse para hacer llamados a API que utilicen SOAP.
    8. Bokeh: Librería que permite hacer animaciones interactivas para aplicaciones web y de escritorio. A diferencia de otras librerías de visualización como Matplotlib, esta se centra en ofrecer gráficos dinámicas que pueden ser usados por el usuario para explorar y manipular los datos de forma directa en la visualización.
    9. Folium: Librería que permite utilizar datos geográficos para crear mapas dinámicos y personalizables, además se integra de forma fácil con otras librerías de análisis de datos geoespaciales, como GeoPandas. Permite además la exportación a código HTML para que los mapas sean fácilmente integrado en sitios web.

```python
# variable de tipo diccionario donde el key es el nombre y el value es la tupla con la información de la persona.
people = {
        "Jenny": (26,39,1.48,True) #Cada tupla contiene cuatro valores: Edad, Peso, Estatura, Estudia
        ,"Hilda": (63,60,1.56,False)
        ,"Felipe": (34,75,1.75,False) 
        ,"Andrés": (32,53,1.70,False)
        ,"Kevin": (37,58,1.68,True)
        ,"Alejandro": (43,68,1.78,False)
        ,"Pedro": (62,65,1.75,False)
        ,"Paola": (30,54,1.63,True)
    }


def get_average_list(list): # función que recibe como variable una lista
    if(isinstance(list[0], bool)): # si la variable contiene boleanos
        # sumar 1 si el valor es true, 0 si es false, luego divide sobre el total de datos para obtener el promedio
        return sum([(1 if item else 0) for item in list])/len(list)
    # si la lista no es de booleanos, suma los valores y divide sobre la longitud de la lista para sacar el promedio
    return sum(list)/len(list)

list_names = ["age","weight","height","study"] # lista para identificar el orden de los valores en la tupla
for index in range(4): # bucle para construir cada una de las 4 listas
  current_list=[]
  for key, value in people.items(): # bucle para leer el key y el value del diccionario
    current_list.append(value[index]) # agregar a la lista el valor correspondiente de la tupla
  average = get_average_list(current_list) # obtiene el promedio usando la función creada
  print(f"average of {list_names[index]}: {average}") # muestra el promedio calculado para la actual lista.
```
