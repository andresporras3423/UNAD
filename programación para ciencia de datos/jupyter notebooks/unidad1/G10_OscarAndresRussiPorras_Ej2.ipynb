    En términos algebráicos se puede expresar que resolver un sistema lineal de ecuaciones e encontrar x de la función:

$$
Ax=b
$$

Donde A es una matriz, x es la columna de las incógnitas y b es la columna vector dee términos independientes. Existen muchos métodos para encontrar la solución, una forma es invirtiendo la matriz, mediante sustitución,y también mediante el método crammer. En este último método vamos sustituyendo una de las columnas de A por la del vector b, encontramos la determinante de esta nueva matriz y la dividimos por la determinante de la matriz A, y repetimos para cada columna y así econtramos todas las incógnitas. Ejemplo, teniendo el siguiente sistema de ecuaciones:

$$
2x + y = 7
\\
x + 3y = 6
$$
para x tenemos que:
$$
\det 
\begin{bmatrix}
7 & 1 \\
6 & 3 \\
\end{bmatrix}
\div
\det
\begin{bmatrix}
2 & 1 \\
1 & 3 \\
\end{bmatrix}
= 
\frac{7 \times 3 - 1 \times 6}{2 \times 3 - 1 \times 1} 
= 
\frac{15}{5}
=
3
$$
Y para y tenemos:
$$
\det 
\begin{bmatrix}
2 & 7 \\
1 & 6 \\
\end{bmatrix}
\div
\det
\begin{bmatrix}
2 & 1 \\
1 & 3 \\
\end{bmatrix}
= 
\frac{2 \times 6 - 7 \times 1}{2 \times 3 - 1 \times 1} 
= 
\frac{5}{5}
=
1
$$
por lo tanto:
$$
x=3, y=1
$$

    Los valores singulares de una matriz permiten entender características importantes el número de condición o el factor de estiramiento. Para encontrar los valores singulares de una matriz A se pueden seguir los siguientes pasos:
1) Obtenemos la siguiente matriz $$ A^{T}A− λI $$
donde I es la matriz identidad y λ es la variable eigenvalue.
2) obtener la determinante de la matriz del paso anterior.
3) deducir los dos eigenvalue.
4) sacar la raiz cuadrado de los eigenvalue.
Ejemplo: teniendo la siguiente matriz:
$$
A=
\begin{bmatrix}
1 & 1 \\
2 & -1 \\
\end{bmatrix}
$$
multiplicamos la transpuesta por esta matriz:
$$
A^{T}A = \begin{bmatrix}
5 & -1 \\
-1 & 2 \\
\end{bmatrix}
$$
le restamos λI:
$$
\begin{bmatrix}
5 & -1 \\
-1 & 2 \\
\end{bmatrix}-λI=
\begin{bmatrix}
5-λ & -1 \\
-1 & 2-λ \\
\end{bmatrix}
$$
Sacamos la determinante:
$$
\det
\begin{bmatrix}
5-λ & -1 \\
-1 & 2-λ \\
\end{bmatrix}
=(5-λ)(2-λ)-(-1)(-1)=λ^{2}-7λ+9
$$
Resolviendo la ecuación cuadrática obtenemos los dos eigenvalues:
$$
λ_1 = 5.30277564, 
λ_2 = 1.69722436
$$
los valores singulares de la matriz serían las raices cuadradas de estos dos valores:
$$
σ_1 = 2.30277,
σ_2 = 1.30277
$$
que son los mismos que obtendríamos en la cuarta entrada de la tupla que retornaría np.linalg.lstsq al aplicarla sobre esta matriz.

<h3>construya la siguiente matriz como un arreglo de numpy:</h3><br>
    $$
A =
\begin{bmatrix}
13 & 1 \\
14 & 1 \\
15 & 1 \\
16 & 1 \\
17 & 1 \\
18 & 1 \\
\end{bmatrix}
$$

```python
import numpy as np
A = np.array([[13, 1], [14, 1], [15, 1], [16, 1], [17, 1], [18, 1]])
print(A)
```

<h3>Construya el siguiente vector como arreglo de numpy:</h3><br>
$$
b =
\begin{bmatrix}
16 \\
24 \\
45 \\
57 \\
75 \\
102 \\
\end{bmatrix}
$$

```python
b = np.array([16,24,45,57,75,102])
print(b)
```

<h3>Calcule el siguiente vector x, usando operaciones de matrices (dot, transpose, inv) de numpy.linalg:</h3><br>
   $$
x=(A^{T}A)^{-1}A^{T}b
$$ 

```python
transpose_A= A.T # transpose of A
TA_dot_A = np.dot(transpose_A, A) # multiply tranpose of A and A
inv_TA_dot_A =  np.linalg.inv(TA_dot_A) # inverse previous result
TA_dot_b = np.dot(transpose_A, b) # multiply tranpose of A and b
x_matrix  = np.dot(inv_TA_dot_A, TA_dot_b)
print(x_matrix)
```

<h3>y calcule ‖𝐴𝑥−𝑏‖ utilizando la función numpy.linalg.norm.</h3><br>

```python
A_x_dot = np.dot(A, x_matrix) # multiply A and x matrix
# print(A_x_dot)
diff_A_x_dot_And_b = A_x_dot - b # get matrix difference
norm_final_matrix = np.linalg.norm(diff_A_x_dot_And_b) # get norm of the previous matrix
print(norm_final_matrix)
```

<h3>Solucione el siguiente sistema de ecuaciones, mediante la función numpy.linalg.solve (de ser necesario, investigue la función en la documentación de numpy):</h3><br>
   $$
(A^{T}A)x=A^{T}b
$$ 
    

```python
x_solution = np.linalg.solve(TA_dot_A, TA_dot_b)
print(x_solution)
```

<h3> y calcule ‖𝐴𝑥−𝑏‖ utilizando la función numpy.linalg.norm.</h3><br>

```python
norm_final2 = np.linalg.norm(np.dot(A,x_solution)-b) # calculate again norm using the matrix calculated in the previous cell
print(norm_final2)
```

<h3>Solucione el siguiente sistema de ecuaciones, mediante la función numpy.linalg.lstsq (de ser necesario, investigue la función en la documentación de numpy):</h3><br>
   $$
Ax=b
$$ 

```python
solved_system = np.linalg.lstsq(A, b, rcond=None) # solve system Ax = b
print(solved_system)
```

<h3>y calcule ‖𝐴𝑥−𝑏‖ utilizando la función numpy.linalg.norm.</h3><br>

```python
norm_final3 = np.linalg.norm(np.dot(A,solved_system[0])-b) # calculate again norm using the second entry of the tuple gotten with the lstsq method
print(norm_final3)
```

        x \approx (A^{T}A)^{-1}A^{T}b

        $$
        A^{-1}b = (A^{T}A)^{-1}A^{T}b
        $$
