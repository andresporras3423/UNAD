<h3>Marco teórico</h3><br>
<lu>
    <li>Una gráfica de puntos de dispersión (conocido como scatter plot en inglés) permite mostrar valores de dos variables para un conjunto de datos. Permiten identificar una posible relación entre las dos variables o detectar anomalías en el conjunto de datos.</li>
    <li>Un histograma es una gráfica de distribución de conjunto de datos, permite obtener una estimación de la densidad de probabilidad de una variable continua. Para hacer un histograma el primer paso es dividir todo el rango de valores en una serie de intervalos y luego contar cuantos datos caen en cada intervalo. Finalmente se presenta cada intervalo con una barra cuya altura representa la cantidad de datos en el intervalo. </li>
    <li>La distribución normal, también llamada distribución gaussiana, es un tipo de distribución que se caracteríza por su forma de campana (vista en un plano cartesiano) y está definida por dos parámetros, la media y la desviación estandar. Es un tipo de distribución muy popular por sus múltiples aplicaciones en diversos escenarios donde los datos se distribuyen de forma simétrica alrededor de la media.</li>
    <li>La distribución uniforme es un tipo de distribución donde todos los valores tienen la misma probabilidad de ocurrir. Este tipo de distribución puede ser discreta o continua.</li>
    <li>plot.scatter es una función de la biblioteca matplotlib que permite crear gráficos de dispersión, útil para visualizar y analizar la relación entre dos variables. Por ejemplo "matplot.pyplot.scatter([1,2,3,4], [1,4,9,16], color='blue', label='cuadratic function')" permite graficar los primeros cuatros enteros positivos de la función: $$ x^2 $$ En color azul y con el label "cuadratic function"</li>
    <li> La función linspace de numpy se usa para crear una lista de números equiespaciados entre el rango definido. Por ejemplo: "np.linspace(-10, 10, 100)" crea una lista con la serie: -10, -9.8, -9.6, -9.4.... 10.</li>
    <li>El método log de numpy permite obtener el logaritmo natural de un número o de cada número de un arreglo de números. Por ejemplo "numpy.log([10, math.e])" retornaría el arreglo [2.3025, 1]</li>
    <li>La función plot de matplotlib.pyplot sirve para graficar funciones lineales o de puntos. Ejemplo: "matplotlib.pyplot.plot(numpy.linspace(0, 10, 100), numpy.linspace(0, 20, 100),color='blue', label='y=2x')" permite graficar en una gráfica lineal la función y=2x usando una línea de color azul.</li>
    <li>La función random.normal de numpy permite generar una muestra aleatoria de distribución normal usando los parámetros  de media y desviación estandar. Ejemplo: "numpy.random.normal(loc=0.0, scale=0.5)" genera una muestra aletatoría con distribución normal, desviación estandar de 0.5 y media de 0.</li>
        <li>La función random.uniform de numpy genera una muestra uniforme aleatoria dentro del rango especificado. Ejemplo: "numpy.random.uniform(low=0.0, high=1.0)" genera una muestra aleatoria con distribución uniforme y valores entre 0 y 1.</li>
    <li>La función hist de matplotlib.pyplot sirve para crear histogramas. Ejemplo: "plt.hist([18,12,19,2,90,32], bins=3, color='blue', edgecolor='black')" generará un histograma de 3 columnas, la primera columna es de altura 4 (correspondiente a los valores: 18, 12, 19 y 2) y altura uno para la segunda y tercera columna (correspondientes a los valores 32 y 90 respectivamente). Además, las columnas son puestas de azul y los bordes entre columnas en color negro.</li>
</lu>



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# función para crear lista de puntos en el rango especificado
def f(func_, label_, color_, min_range, max_range, ln_):
    x = np.linspace(min_range, max_range, 500) # 500 points between
    
    y = func_(x)
    plt.plot(x, y, label=label_, color=color_,linestyle=ln_)
# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'A': [13,14,15,16,17,18],
    'B': [16,24,45,57,75,102]
})

# Gráfica de dispersión usando datos del DataFrame
df.plot.scatter(x='A', y='B', label= 'pairing lists', color='green')

# función lineal
fun= lambda z: (17*z)-210.3333 # 𝟏𝟕𝒙−𝟐𝟏𝟎.𝟑𝟑3𝟑
f(fun, "y = 17z-210.3333","yellow",13,18,"-")

# función exponencial
fun2= lambda z: 0.1452*(math.e**(0.369*z))
f(fun2, "y = 0.1452e^(0.369z)","blue",13,18,"-")

# función logarítmica
fun3= lambda z: (259.49*np.log(z))-656.46
f(fun3, "y = 259.49ln(z)-656.4","red",13,18,"dashed")

plt.title("graph 1") # título gráfica
plt.xlabel("x") # label eje x
plt.ylabel("y") # lavel eje y
plt.grid(True) # agregar grilla a la gráfica
plt.legend() # agregar leyenda 
plt.show() # mostrar gráfica
```

```python
normal= lambda mu, sigma: np.random.normal(mu, sigma, 700) # función para la distribución normal
uniform = lambda mi, ma: np.random.uniform(mi, ma, 700) # función para la distribución uniforme

x1 = normal(7, 0.1) # media: 7, desviación estandar 0.1
y1 = normal(5, 0.1) # media: 5, desviación estandar 0.1
x2= uniform(min(x1),max(x1)) # distribución uniforme entre el mínimo y el máximo valor de x1
y2= uniform(min(y1),max(y1)) # distribución uniforme entre el mínimo y el máximo valor de y1
plt.scatter(x1, y1, color='blue', label='(x1, y1)', alpha=0.5) # diagrama de dispersión con las dos variables de distribución normal

plt.scatter(x2, y2, color='red', label='(x2, y2)', alpha=0.5) # diagrama de dispersión con las dos variables de distribución uniforme


plt.title('Graph 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
excel_file_path=r"Anexo 1 - Base de Datos EXCEL.xlsx"
sheet_name = "BASE DE DATOS" # nombre de la hoja
skiprows = 3           # filas iniciales a ignorar
usecols = 'A:M'        # solo leer columnas de la A a la M
header = 0             #  usar primera fila como headers

df1 = pd.read_excel(excel_file_path, sheet_name=sheet_name, skiprows=skiprows, usecols=usecols, header=header)
tiempo_motor_on = df1['Tiempo motor ON (h)'] # guardar en una variable los valores de la columna 'Tiempo motor ON (h)'
tiempo_en_movimiento = df1['Tiempo en Movimiento (h)'] # guardar en una variable los valores de la columna 'Tiempo en Movimiento (h)'

# Crear los histogramas
plt.hist(tiempo_motor_on, alpha=0.5, label='Tiempo motor ON', color='blue')
plt.hist(tiempo_en_movimiento, alpha=0.5, label='Tiempo en Movimiento', color='red')

# Añadir título, etiquetas y leyenda
plt.title('Histogramas de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# calculamos el error medio cuadrático para ver cuál función se ajusta mejor a los puntos de dispersión
def error_calc(func_type, func_):
    df = pd.DataFrame({
        'A': [13,14,15,16,17,18],
        'B': [16,24,45,57,75,102]
    })
    suma=0
    for i in range(6):
        suma+=(df['B'][i]-func_(df['A'][i]))**2
    print(f"error función {func_type}: {suma/6}")

fun3= lambda z: (259.49*math.log(z))-656.46
error_calc("logarítmica",fun3)
fun2= lambda z: 0.1452*(math.e**(0.369*z))
error_calc("exponencial", fun2)
# Generate x values
fun= lambda z: (17*z)-210.3333 # 𝟏𝟕𝒙−𝟐𝟏𝟎.𝟑𝟑3𝟑
error_calc("lineal",fun)
```
