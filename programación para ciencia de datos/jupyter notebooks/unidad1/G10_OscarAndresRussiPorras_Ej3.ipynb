```python
import numpy as np


def f(func_, label_, color_, title_, min_range, max_range, ln_):
    x = np.linspace(min_range, max_range, 500) # 500 points between -10 and 10
    y = func_(x)
    plt.plot(x, y, label=label_, color=color_,linestyle=ln_)
    plt.title(title_)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()
    plt.show()
# Define the function
#def f(x):
#    return (17*x)-210.3333

# Generate x values
fun= lambda z: (17*z)-210.3333 # 𝟏𝟕𝒙−𝟐𝟏𝟎.𝟑𝟑3𝟑
f(fun, "y = 17z-210.3333","yellow","Graph of y = 17z-210.3333",13,18,"-")
```

```python
import math

fun= lambda z: 0.1452*(math.e**(0.369*z))
f(fun, "y = 0.1452e^(0.369z)","blue","Graph of y = 0.1452e^(0.369z)",13,18,"-")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def f(func_, label_, color_, min_range, max_range, ln_):
    x = np.linspace(min_range, max_range, 500) # 500 points between -10 and 10
    y = func_(x)
    plt.plot(x, y, label=label_, color=color_,linestyle=ln_)
# Define the function
#def f(x):
#    return (17*x)-210.3333

# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'A': [13,14,15,16,17,18],
    'B': [16,24,45,57,75,102]
})

# Gráfica de dispersión usando datos del DataFrame
df.plot.scatter(x='A', y='B', label= 'pairing lists', color='green')

# Generate x values
fun= lambda z: (17*z)-210.3333 # 𝟏𝟕𝒙−𝟐𝟏𝟎.𝟑𝟑3𝟑
f(fun, "y = 17z-210.3333","yellow",13,18,"-")

fun2= lambda z: 0.1452*(math.e**(0.369*z))
f(fun2, "y = 0.1452e^(0.369z)","blue",13,18,"-")

fun3= lambda z: (259.49*np.log(z))-656.46
f(fun3, "y = 259.49ln(z)-656.4","red",13,18,"dashed")

plt.title("graph 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
excel_file_path=r"Anexo 1 - Base de Datos EXCEL.xlsx"
sheet_name = "BASE DE DATOS" # Specify the sheet name
skiprows = 3           # Skip the first three rows
usecols = 'A:M'        # Only read columns A to M
header = 0             # Use the first row as column headers

df1 = pd.read_excel(excel_file_path, sheet_name=sheet_name, skiprows=skiprows, usecols=usecols, header=header)
tiempo_motor_on = df1['Tiempo motor ON (h)']
tiempo_en_movimiento = df1['Tiempo en Movimiento (h)']

# Crear los histogramas
plt.hist(tiempo_motor_on, alpha=0.5, label='Tiempo motor ON', color='blue')
plt.hist(tiempo_en_movimiento, alpha=0.5, label='Tiempo en Movimiento', color='red')

# Añadir título, etiquetas y leyenda
plt.title('Histogramas de Tiempo motor ON y Tiempo en Movimiento')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
 df = pd.DataFrame({
    'Tiempo motor on': tiempo_motor_on,
    'Tiempo en movimiento': tiempo_en_movimiento
})
# Gráfica de dispersión usando datos del DataFrame
df.plot.scatter(x='Tiempo motor on', y='Tiempo en movimiento', color='green')
plt.show()
```
