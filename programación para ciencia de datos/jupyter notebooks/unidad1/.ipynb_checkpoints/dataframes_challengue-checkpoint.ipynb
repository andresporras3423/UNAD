```python
import pandas as pd
excel_file_path=r"Anexo 1 - Base de Datos EXCEL.xlsx"
sheet_name = "BASE DE DATOS" # Specify the sheet name
skiprows = 3           # Skip the first three rows
usecols = 'A:M'        # Only read columns A to M
header = 0             # Use the first row as column headers

df1 = pd.read_excel(excel_file_path, sheet_name=sheet_name, skiprows=skiprows, usecols=usecols, header=header)
df1
# df1.describe(include=[int])
# df1.info(show_counts=True, memory_usage=True, verbose=True)
```

```python
csv_file_path = r"Anexo 2 - Base de Datos CSV.csv"
df2 = pd.read_csv(csv_file_path)
df2
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()

df3 = pd.DataFrame(data=cancer_data.data, columns= cancer_data.feature_names)      
df3
```

```python
df_dict = {'tabla excel': df1, 'tabla csv': df2, 'tabla cancer breast': df3}
df_combined = pd.concat([df1, df2, df3], keys=['tabla excel', 'tabla csv', 'tabla cancer breast'])
print(df_combined)
```

```python
def extract_column_headers(panel_df):
    for name, df in panel_df.items():
        print(f'dataframe {name}:')
        print("\n".join(df.columns.tolist()))
        print("\n")
extract_column_headers(df_dict)
```

```python
def extract_dataframe_sizes(panel_df):
    for key, group_df in panel_df.items():
        print(f"nombre del dataframe: {key}")
        print(f"dimensión: ")
        print(f"filas: {len(group_df)}, columnas: {len(group_df.columns)}\n")

# Usamos la función en el DataFrame concatenado (suponiendo que se llama df_combined)
headers = extract_dataframe_sizes(df_dict)
```
