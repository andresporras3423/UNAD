<h3>Introducción</h3>
<ul>
    <li><b>¿Qué es REST web service y cómo se utilizan?:</b>
    REST (cuyas siglas significan Representational State Transfer) es un estilo arquitectonico para compartir recursos en la web. Estos recursos son compartidos a través de una uri, por ejemplo, un REST web service de empleados de una compañía podría ser https://company.com/empleados/123 donde 123 representa el identificador de un empleado, entonces esta URI podría usarse para obtener la información del empleado 123. Además de la URI, las peticiones de recursos en un rest web service tiene las siguientes características:
        <ol>
            <li> Operaciones CRUD: Identifica el tipo de operaciones que se pueden hacer sobre los recursos, por ejemplo crear, leer, eliminar o actualizar recursos. </li>
        <li> stateless: cada petición que el cliente hace sobre los recursosdebe contener toda la información necesaria para procesar la petición, de forma que el servicio web no guarda información entre peticiones. </li>
            <li> formato de datos: los datos son transmitidos en formato JSON o XML.</li>
        <li> código de estado HTTP: los servicios RESTFUL usan códigos para informar el resultado de la petición, por ejemplo 200 indica éxito, 404 not found, o 500 es internal server error.</li>
         <li> arquitectura cliente-servidor en la arquitectura REST, el cliente y el servidor son entidades diferentes que se comunican entre sí. El cliente es el que hace la petición de recursos y el servior es quien ofrece los recursos.</li>
        <li> cacheable: el cliente puede almacenar copias de las respuestas del servidor, permitiendo mejorar la velocidad y reducir la carga.</li>
        </ol><br>
        <span> desde el lado del cliente, se pueden hacer peticiones REST complejas a través de herramientas como POSTMAN.</span> 
    </li><br>
    <li>
        <b>¿Qué es una RESTful API?:</b> es una interfaz para programación de aplicaciones que siguen los principios de REST. El concepto "API"  indica un conjunto de reglas y mecanismos para la comunicación entre un software y otro. El diseño y planificación de cómo los sistemas deberían comunicarse o que recursos deberían estar disponibles para ciertos roles de usuarios, esto corresponde al diseño del API. Para la creación de un API se pueden seguir los siguientes pasos:
        <ol>
        <li>identificar los recursos: esto es las entidades que se ofrecerán, por ejemplo: usuarios, empleados, clientes, etc.</li>
         <li>definir las URIs: definir la estructura de URI que identificará cada recurso, por ejemplo "/empleados" para los empleados.</li>  
        <li> utilizar los métodos GET, POST, PUT, y DELETE (entre otros) para interactuar con los recursos.</li>
        <li> utilizar códigos HTTP para identificar los resultados de las peticiones a los recursos.</li>
            <li> agregar mecanismos de seguridad y autenticación, por ejemplo tokens y OAuth.</li>
        <li> para la creación de servicios web RESTFUL se pueden usar frameworks como Express para node.js, Flask para Python, Spring Boot para Java, NET Framework para C#, entre otros.</li>
        </ol>
    </li><br>
    <li><b>¿Qué es JSON?:</b> JSON (cuyas iniciales significan JavaScript Object Notation)es un formato ligero utilizado para el intercambio para compartir datos. Aunque se originó inspirado en la notación de objetos de JavaScript es un formato independiente del lenguaje y puede ser utilizado con cualquier languaje de programación. JSON tiene las siguientes características:
    
 <ol>
         <li>Es un lenguaje textual, diseñado para ser fácil de escribir y leer para los humanos y fácil de compartir entre sistemas.</li>
         <li>Estructura simple basada en dos subestructuras fundamentales, primero colección de pares llave/valor, y en segundo lugar una lista ordenada de valores.</li>
         <li>Soporta la siguiente lista de valores: objectos, arreglos, números, cadenas, booleanos y el valor null.</li>
         <li>Se ha convertido en el estándar para los APIs gracias a su simplicidad y legibilidad.</li>
         <li>Independencia de lenguaje, aunque fue inspirado en los objetos de JavaScript su estructura simple ha facilidato su implementación en todos los lenguajes populares.</li>
        </ol>
El siguiente es un ejemplo de un objeto json:
        
{
    "nombre": "Oscar",
    "edad": 32,
    "estudiante": true,
    "hobbies": ["leer", "música", "netflix"],
    "residencia": {
        "país": "Colombia",
        "ciudad": "Bogotá",
        "barrio": "7 de Agosto"
    }
}
        
Este ejemplo muestra un objeto JSON con la llave nombre que es un string, edad que es un número, estudiante que es booleano, hobbies que es una lista de strings, y residencia que es un objecto con tres llaves que son strings.
</li>
</ul>

<h3><ul> Desarrolle un código de python que le permita obtener los datos
de las naves de SpaceX, desde la API spacexdata v4:
https://api.spacexdata.com/v4/ships, tenga en cuenta que debe
usar el paquete requests, funciones:
    <li> request.get</li>
    <li> Response.json</li>
<li> De ser necesario, puede revisar su documentación en:
    https://requests.readthedocs.io/en/latest/api/?requests.Request</li><ul></h3>

```python
import requests
import pandas as pd

url = "https://api.spacexdata.com/v4/ships"
response = requests.get(url)
data = response.json()
print(data)
```

<h3>Cargue el json de la respuesta del request anterior como un
DataFrame de pandas (ayuda: función pd.DataFrame.from_dict).</h3>

```python
df = pd.DataFrame.from_dict(data)
print(df)
```

<h3>Determine el total de naves y cuántas están activas.</h3>

```python
#print(df.iloc[0]['roles'])
#print(len([rol for rol in df.iloc[0]['roles'] if 'ship' in rol.lower()])>0)
df['is_ship']=df['roles'].apply(lambda roles: any('ship' in rol.lower() for rol in roles))
ships = df[df['is_ship']]
total_ships = len(ships)
print(f"Total ships are: {total_ships}")
total_active = len(ships[ships['active']==True])
print(f"Total active ships are: {total_active}")
```

<h3>Visualice el histograma de los pesos de las naves en kg</h3>

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
n, bins, _ = plt.hist(df[df['mass_kg']>0]['mass_kg'], bins=[0, 1000000, 2000000, 3000000, 4000000, 5000000], edgecolor="k", label='ship weight')

# Añadir las frecuencias encima de cada columna
for i in range(len(n)):
    plt.text((bins[i] + bins[i+1]) / 2, n[i] + 0.5, str(int(n[i])), ha='center', va='center')


formatter = mticker.ScalarFormatter(useMathText=False)
formatter.set_scientific(False)
plt.gca().xaxis.set_major_formatter(formatter)
# Añadir título, etiquetas y leyenda
plt.title('Histogramas del peso de las naves en kilogramos')
plt.xlabel('weight(in kgs)')
plt.ylabel('frequency')
plt.legend()
plt.show()
```
