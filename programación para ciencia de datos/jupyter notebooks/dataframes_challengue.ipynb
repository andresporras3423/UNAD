```python
import pandas as pd
from sklearn.datasets import fetch_openml
excel_file_path=r"Y:\Documents\UNAD\programación para ciencia de datos\Anexo 1 - Base de Datos EXCEL.xlsx"
sheet_name = "BASE DE DATOS" # Specify the sheet name
skiprows = 3           # Skip the first three rows
usecols = 'A:M'        # Only read columns A to C
header = 0             # Use the first row as column headers

df1 = pd.read_excel(excel_file_path, sheet_name=sheet_name, skiprows=skiprows, usecols=usecols, header=header)

csv_file_path = r"Y:\Documents\UNAD\programación para ciencia de datos\Anexo 2 - Base de Datos CSV.csv"
df2 = pd.read_csv(csv_file_path)

# Load the Boston Housing dataset from UCI Machine Learning Repository
boston = fetch_openml(data_id=506, parser='auto')

# Access the data, target values, feature names, and other metadata
df3 = boston.data       # The feature matrix
df_combined = pd.concat([df1, df2, df3], keys=['df1', 'df2', 'df3'])
# print(df_combined)
def extract_column_headers(panel_df):
    # Obtenemos los niveles únicos del primer nivel del índice (que representa el DataFrame original)
    # print(panel_df.index)
    print(panel_df.index.get_level_values(1))
    dataframe_labels = panel_df.index.get_level_values(0).unique()

    # Diccionario para guardar los encabezados de cada DataFrame original
    headers_dict = {}
    
    for label in dataframe_labels:
        # Filtramos el panel para obtener solo las filas de un DataFrame específico
        subset = panel_df.loc[label]

        # Guardamos los encabezados de las columnas que no son totalmente NaN
        headers_dict[label] = subset.columns[~subset.isna().all()].tolist()
        
    return headers_dict

# Usamos la función en el DataFrame concatenado (suponiendo que se llama df_combined)
headers = extract_column_headers(df_combined)

# Imprimimos los resultados
for df_label, cols in headers.items():
    print(f"Encabezados de {df_label}:")
    print(cols)
    print("\n")
```

```python
import pandas as pd

# Datos
data = {
    'Ventas': [100, 200, 150, 300],
    'Ingresos': [5000, 4000, 5500, 2000]
}

# MultiIndex: Tuples que representan las combinaciones de (Región, Categoría)
index = pd.MultiIndex.from_tuples([
    ('Norte', 'Electrónicos'),
    ('Norte', 'Ropa'),
    ('Sur', 'Electrónicos'),
    ('Sur', 'Alimentos')
], names=['Región', 'Categoría'])

# Crear DataFrame con MultiIndex
df = pd.DataFrame(data, index=index)

# Mostrar DataFrame
print(df)
```
